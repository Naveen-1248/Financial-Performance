```python
 # Import libraries
import pandas as pd
import numpy as np

# -----------------------------
# 1. CREATE SAMPLE DATASET
# -----------------------------
data = {
    "Invoice_ID": [101, 102, 103, 104, 105, 106],
    "Customer_Name": ["ABC Ltd", "XYZ Pvt", "ABC Ltd", "LMN Corp", "XYZ Pvt", "PQR Ltd"],
    "Invoice_Date": ["2024-01-01", "2024-01-05", "2024-01-10", "2024-01-15", "2024-01-20", "2024-01-25"],
    "Due_Date": ["2024-01-10", "2024-01-15", "2024-01-20", "2024-01-25", "2024-01-30", "2024-02-05"],
    "Payment_Date": ["2024-01-09", "2024-01-20", None, "2024-01-24", None, "2024-02-10"],
    "Invoice_Amount": [5000, 7000, 6000, 8000, 7500, 9000]
}

df = pd.DataFrame(data)

# Convert date columns to datetime
df["Invoice_Date"] = pd.to_datetime(df["Invoice_Date"])
df["Due_Date"] = pd.to_datetime(df["Due_Date"])
df["Payment_Date"] = pd.to_datetime(df["Payment_Date"])

# -----------------------------
# 2. DATA CLEANING
# -----------------------------
# Fill missing Payment_Date with today's date (for analysis)
df["Payment_Date"].fillna(pd.Timestamp.today(), inplace=True)

# -----------------------------
# 3. CREATE NEW COLUMNS
# -----------------------------
# Payment status
df["Payment_Status"] = np.where(df["Payment_Date"] <= df["Due_Date"], "Paid On Time", "Late")

# Days taken to pay
df["Payment_Days"] = (df["Payment_Date"] - df["Invoice_Date"]).dt.days

# -----------------------------
# 4. KPI CALCULATIONS
# -----------------------------
total_revenue = df["Invoice_Amount"].sum()
avg_payment_days = df["Payment_Days"].mean()
late_payments = df[df["Payment_Status"] == "Late"].shape[0]
total_invoices = df.shape[0]

late_percentage = (late_payments / total_invoices) * 100

# -----------------------------
# 5. CUSTOMER ANALYSIS
# -----------------------------
customer_summary = df.groupby("Customer_Name").agg({
    "Invoice_Amount": "sum",
    "Invoice_ID": "count",
    "Payment_Days": "mean"
}).rename(columns={
    "Invoice_Amount": "Total_Amount",
    "Invoice_ID": "Total_Invoices",
    "Payment_Days": "Avg_Payment_Days"
}).sort_values(by="Total_Amount", ascending=False)

# -----------------------------
# 6. PRINT RESULTS
# -----------------------------
print("📊 KPI RESULTS")
print("Total Revenue:", total_revenue)
print("Average Payment Days:", round(avg_payment_days, 2))
print("Late Payment %:", round(late_percentage, 2), "%")

print("\n📊 Customer Summary")
print(customer_summary)

# -----------------------------
# 7. INSIGHTS
# -----------------------------
print("\n📌 INSIGHTS:")
if late_percentage > 30:
    print("- High number of late payments affecting cash flow.")

top_customer = customer_summary.index[0]
print(f"- Top revenue generating customer: {top_customer}")
```
